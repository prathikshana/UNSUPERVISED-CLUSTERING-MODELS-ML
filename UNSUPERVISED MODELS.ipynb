# dimensionality reduction using PCA such that the 95% of the variance is explained


```python
# scale the data

from sklearn.preprocessing import StandardScaler
scale =  StandardScaler()
datascale = scale.fit_transform(data)
datascale
```

```python
#pca model
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(datascale)
```

```python
# eigen value and eigen vector 
cov_m = np.cov(datascale.T)  #covariance matrix
eigen_value, eigen_vector = np.linalg.eig(cov_m) # eigen info
```

```python
print(eigen_value)
```

```python
print(eigen_vector)
```

```python
exp = pca.explained_variance_ratio_
print(exp)
print(exp.shape,'shape')
```

```python
plt.figure(figsize=(6, 4))

plt.bar(range(0,17), exp, alpha=0.9,align='center',
            label='individual explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
```

```python
# finding pca componnets
np.sum(exp[0:12])
```

## Conclusion of PCA:
SO BY USING 13 PCA COMPONENTS TO REDUCE THE DATA SET DIMENSION , WHICH CONTAINS 95 OR ABOVE INFORMATION OF DATA POINTS IN DATA SPACE WE WILL BE ABLE TO OBTAIN OPTIMAL ACCURACY

# K-MEANS CLUSTERING USING ELBOW METHOD

```python
#kmeans model
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kfit = kmeans.fit(data)
```

```python
clusterdata = data.copy()
clusterdata['dpredict'] = kfit.predict(data)
clusterdata
```

```python
sb.scatterplot(x=clusterdata['PURCHASES'],y=clusterdata['PURCHASES_FREQUENCY'],hue=clusterdata['dpredict'],data=clusterdata.head(100))
```

##### elbow method

```python
from sklearn import preprocessing
kscale = preprocessing.scale(data)
kscale
```

```python
elbow = []
for i in range (1,10):
    kmeanse = KMeans(n_clusters=i)
    kmeanse.fit(kscale)
    elbow.append(kmeanse.inertia_)
plt.plot(range(1,10),elbow)
plt.grid()
    
```

From elbow line plot it states that lines curve downs at 2 , so the number of clusters for this dataset is 2

```python
kmean = KMeans(n_clusters = 2)
kmfit = kmean.fit(kscale)
kmfit
```

```python
kmdata = data.copy()
kmdata['predict'] = kmfit.predict(kscale)
kmdata
```

```python
sb.scatterplot(x='PURCHASES_FREQUENCY',y='PURCHASES',hue='predict',data=kmdata.head(100))
```
